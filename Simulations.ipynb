```python
''' useful short functions'''
# error function
def error_func(a,b):
    if abs(a-b)<10**(-14):
        return True
    else:
        return False

# function to calculate expected value given values and defenses.
def expected_v(v, d):
    return v*exp(-d)

expected_v = np.vectorize(expected_v)
```

```python
"""
Function
--------

create_V_expo

Parameters
----------
num_Vs: integer
    Number of targets you wish to protect. (Length of V).
    
Lam: float
    Lamdba to be used for the exponential distribution
    
number_of_vectors: integer
    Number of vectors with len(num_Vs) targets you want.
   
Returns
--------
Vectors: list of lists
List of length number_of_vectors. Each element in the list is itself a list 
containing length(num_Vs) targets with their corresponding value.

"""
########## get V's (True value known to terrorists) ###############
def create_V_expo(num_Vs, Lam, number_of_vectors):
    Vectors = []
    for i in xrange(number_of_vectors):
        V = np.zeros(num_Vs) #array of values
        for element in xrange(num_Vs):
            V[element] = random.expovariate(Lam)
        V = np.sort(V)[::-1] #order targets from most to least import.   
        Vectors.append(V)
    return Vectors
```

```python
# *** Note, this function assumes that V's are ordered!!!!!
"""
Function
--------

Push_down_gen

Parameters
----------

Vec:
    Vector of current target values (target probabilities?).

Budget: float
    the defense budget we have at our disposal. 
    
Returns
--------

Vector detailing defense units tobe spent on each target

"""
def Push_down_gen(Budget,Vec):
    Def = np.zeros(len(Vec))
    Def_temp = np.zeros(len(Vec))
    
    # cycling through limiting values of targets.
    for v in xrange(1,len(Vec)): 
        
        #cycling all targets with higher values than v.
        for i in xrange(v):

            Def_temp[i] = np.log(Vec[i]/Vec[v])
            
        # Case A: if overbudget, use all budget to lower as much as possible
        # set up problem as a matrix.
        if np.sum(Def_temp) > Budget:    

            a1 = np.ones((1,v)) #v and not v-1 because of 0 indexing
            a2 = np.zeros((v-1,v))
            b = np.zeros(v)
            b[0] = Budget
            
            for stuff in xrange(v-1):
                a2[stuff,0]=1
                a2[stuff,stuff+1] = -1
                b[stuff+1] = np.log(Vec[0]) - np.log(Vec[stuff+1])

            a = np.concatenate((a1, a2), axis=0)
            x = np.linalg.solve(a, b)
            add = np.zeros(len(Vec)-len(x))
            return np.concatenate((x,add), axis=0)
        
        # Case B: if exactly on budget, return array of defenses
        elif np.sum(Def_temp) == Budget: 
            return Def_temp
        
        # Case C: if under budget, continue incrementing defense vector.
        else:                       
            Def = np.copy(Def_temp)
    
    # Case D: reaches all targets and is able to bring down all of them.
    a1 = np.ones((1,len(Vec))) #v and not v-1 because of 0 indexing
    a2 = np.zeros((len(Vec)-1,len(Vec)))
    b = np.zeros(len(Vec))
    b[0] = Budget
    
    for stuff in xrange(len(Vec)-1):
        a2[stuff,0]=1
        a2[stuff,stuff+1] = -1
        b[stuff+1] = np.log(Vec[0]) - np.log(Vec[stuff+1])

    a= np.concatenate((a1, a2), axis=0)
    Def = np.linalg.solve(a, b)

    return Def
```

```python
"""
Function
--------

Losses_alt

Parameters
----------
   
Budget_low: float
     lower limit of the defense budget we have at our disposal. 
    
    
Budget_high: float
     upper limit of the defense budget we have at our disposal. 
    
granularity:
    granularity of our analysis

V: list
    list of current target values
    
Returns
--------

reasonable budget : float

"""
def Losses_alt(Budget_low, Budget_high, granularity, V):
    losses = []
    Var_budget = np.linspace(Budget_low, Budget_high, num = granularity)

    
    for b in Var_budget:
        
        Defense_accurate = Push_down_gen(b,V)
        
        #expected values with perfect information
        expv_perfect_info  = expected_v(V, Defense_accurate)
        
        
        # pick highest target (we assume that's what the terrorist will choose)
        loss = expv_perfect_info[0]
        losses.append(loss)
        
    '''
    # Uncomment if visual representation of reasonable budget is needed
    plt.plot(Var_budget, losses) #plt.plot(x, y)
    plt.plot(Var_budget,Var_budget)
    plt.xlabel("Defense Budget")
    plt.ylabel("Losses in expected value")
    plt.title("Plot of losses with varying budget")
    '''
    for element in xrange(len(losses)):
        if losses[element] - Var_budget[element] <0:
            flips =(Var_budget[element]+Var_budget[element-1])/2 
            return flips
    
    return False #only reaches this statement if is reasonable budget exceeds budget constraint
```

```python
### Given the new budget (gotten from the graph), what is the distribution of defenses/which target is attacked?
"""
Function
--------

indices_of_attacked

Parameters
----------

Adjusted_Budget: float
    The "reasoable" budget a country should spend on defending its tarets.

V: list
    list of potential targets.
    
Returns
--------
list of indices of targets that will be at least partially protected.

"""

def indices_of_attacked(V, Adjusted_Budget):
    defenses = Push_down_gen(Adjusted_Budget,V)
    actual_expv=expected_v(V, defenses)
    indices = [i for i, j in enumerate(actual_expv) if error_func(j,max(actual_expv))]
    return indices #indices of potential targets
```

```python
''' Question: for a given number of targets and a given distribution:
** what is the mean 'reasonable' budget? 
** what is the mean number of targets we can protect? '''

"""
Function
--------

mean_budget

Parameters
----------
Budget_low: float
     lower limit of the defense budget we have at our disposal. 
    
    
Budget_high: float
     upper limit of the defense budget we have at our disposal. 
     
granularity:
    granularity of our analysis

Vectors: list of list
    each element in the list is a list of target values for a given simulation.

Returns
--------
(reasonable_budget, indx_len)

reasonable_budget: list
    list of floats representing the appropriate "reasonable budget" for that draw

indx_len: list
    list of integers of how many targets this adjusted budget allows us to defend

"""

def mean_budget_targets(Budget_low, Budget_high, granularity, Vectors):
    reasonable_budget = []
    indx_len = []
    for v in Vectors:
        Adjusted_Budget = Losses_alt(Budget_low, Budget_high, granularity , v)
        reasonable_budget.append(Adjusted_Budget)
        indx = indices_of_attacked(v, Adjusted_Budget)
        indx_len.append(len(indx))

    return reasonable_budget, indx_len
```

```python
"""
Function
--------

Varying_lambda_fixed_N

Parameters
----------

lambda_min: float
    Minimun value of lambda. must be >0.

lambda_max: float
    Maximun value of lambda. 

lambda_s: float
    Stepsize between lambdas


N_min: int
    Minimun number of potential targets

N_max: int
    Maximum Number of potential targets

S: int
    Stepsize between target samplings.
    
number_of_vectors: int
    Number of vectors per each simulation for given lambda. Default is 50.
    
Returns
--------
budget_mean_list - a vector of vectors .each entry signifies the mean reasonable budget for a given target number and a given lambda.
budget_std_list-  vector of vectors. each entry signifies the mean reasonable budget for a given target number and a given lambda.
indx_mean_list - vector of vectors. each entry signifies the # of protected targets for a given target number and a given lambda.
indx_std_list -  vector of vectors. each entry signifies the # of protected targets for a given target number and a given lambda.


Columns, like budget_mean_list[:,3], give a list with a given lambda varying number of targets.
Rows, like budget_mean_list[1,:],  give a list with a given number of targets, varying lambda.

"""

def Varying_lambda_fixed_N(lambda_min,lambda_max,lambda_s, N_min, N_max, S, number_of_vectors= 100):
    number_of_targets = 1+ np.floor((N_max-N_min)/S) 
    number_of_lambdas = (1/lambda_s)*(lambda_max-lambda_min)+1 # note that a transformation is required because xrange does not accomodate floats and 
                                                               # np.linspace requires granularity, not stepsize.
    #initialize vector of vectors 
    budget_std_list = np.zeros((number_of_targets, number_of_lambdas))
    budget_mean_list = np.zeros((number_of_targets, number_of_lambdas))
    indx_std_list = np.zeros((number_of_targets, number_of_lambdas))
    indx_mean_list = np.zeros((number_of_targets, number_of_lambdas))
    
    for j,i in enumerate(np.linspace(lambda_min,lambda_max, number_of_lambdas)):    
        random.seed(2) # use of seed is for reproducible results. seed chosen at random.
        for pos, target_count in enumerate(xrange(N_min, N_max+1, S)):
            Vectors = create_V_expo(num_Vs = target_count, Lam =i, number_of_vectors = number_of_vectors)
            reasonable_budget, indx_len = mean_budget_targets(Budget_low = 0, Budget_high =10, granularity = 60, Vectors = Vectors)
            plt.show()
            budget_std_list[pos, j] = np.std(reasonable_budget) # std in reasonable budget
            budget_mean_list[pos, j] = np.mean(reasonable_budget) # mean od reasonable budget
            indx_std_list[pos, j] = np.std(indx_len) # std in number of targets we can protect
            indx_mean_list[pos, j] = np.mean(indx_len) # mean of targets we can protect
        
    return budget_mean_list, budget_std_list, indx_mean_list, indx_std_list
    
```

## Now let's perform the analysis for the Power Law Distribution

```python
# where y is a uniform variate, n is the distribution power, x0 and x1 define the range of the distribution, 
# and x is your power-law distributed variate.
# http://mathworld.wolfram.com/RandomNumber.html

"""
Function
--------

create_V_Power

Parameters
----------
num_Vs: integer
    Number of targets you wish to protect. (Length of V).
    

number_of_vectors: integer
    Number of vectors with len(num_Vs) targets you want.
    
n: integer
    the degree of our power series. For example if degree = 3, we're drawing from 1/(x**3)
   
Returns
--------
Vectors: list of lists
List of length number_of_vectors. Each element in the list is itself a list 
containing length(num_Vs) targets with their corresponding value.

"""

def create_V_Power(num_Vs, n , number_of_vectors):
    Vectors = []
    V = np.zeros(num_Vs)
    #random.seed(2) # use of seed is for reproducible results. seed chosen at random.
    for i in xrange(number_of_vectors):
        for j in xrange(num_Vs):
            y = random.uniform(0, 1)
            temp= (-y+1)**((1)/(-n+1))
            V[j] = temp
        X = np.sort(V)[::-1] #order targets from most to least import.
        Vectors.append(X)
    return Vectors
```

```python
"""
Function
--------

Varying_degree_fixed_N

Parameters
----------

degree_min: float
    Minimun value of lambda. must be >1.

degree_max: float
    Maximun value of lambda. 

degree_s: float
    Stepsize between lambdas


N_min: int
    Minimun number of potential targets

N_max: int
    Maximum Number of potential targets

S: int
    Stepsize between target samplings.
    
number_of_vectors: int
    Number of vectors per each simulation for given lambda. Default is 50.
    
Returns
--------
budget_mean_list
budget_std_list
indx_mean_list
indx_std_list


"""
# N - integer. max amount of number of targets. Will examine at intervals of 10.
# 1+(np.floor((N-2)/10)), explain number.

def Varying_degree_fixed_N(degree_min,degree_max, degree_s, N_min, N_max, S, number_of_vectors= 100):
    number_of_targets = 1+ np.floor((N_max-N_min)/S) #calculate granularity from stepsize
    #print "number of targets", number_of_targets
    number_of_degrees = (1/degree_s)*(degree_max-degree_min)+1 # note that a transformation is required because xrange does not accomodate floats and 
                                                               # np.linspace requires granularity, not stepsize.
    #print "number_of_lambdas", number_of_lambdas
    budget_std_list = np.zeros((number_of_targets, number_of_degrees))
    budget_mean_list = np.zeros((number_of_targets, number_of_degrees))
    indx_std_list = np.zeros((number_of_targets, number_of_degrees))
    indx_mean_list = np.zeros((number_of_targets, number_of_degrees))
    
    for j,i in enumerate(np.linspace(degree_min,degree_max, number_of_degrees)):
        print "degree", j,i
    
        random.seed(2) # use of seed is for reproducible results. seed chosen at random.
        for pos, target_count in enumerate(xrange(N_min, N_max+1, S)):
            #print "N", target_count, "pos", pos 
            Vectors = create_V_Power(num_Vs = target_count, n =i, number_of_vectors= number_of_vectors)
            #print Vectors
            
            reasonable_budget, indx_len = mean_budget_targets(Budget_low = 0, Budget_high =10, granularity = 30, Vectors = Vectors)
            plt.show()
            #print "reasonable budget", reasonable_budget
            budget_std_list[pos, j] = np.std(reasonable_budget) # std in reasonable budget
            budget_mean_list[pos, j] = np.mean(reasonable_budget) # mean od reasonable budget
            indx_std_list[pos, j] = np.std(indx_len) # std in number of targets we can protect
            indx_mean_list[pos, j] = np.mean(indx_len) # mean of targets we can protect
        
    return budget_mean_list, budget_std_list, indx_mean_list, indx_std_list
```
